```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#import citi bike data from 01-06/2019 & 01-06/2020 to see
#if there are any effects on rentals from COVID-19

#2019 Data
tripdata_119 = "Resources/JC-201901-citibike-tripdata.csv"
tripdata_219 = "Resources/JC-201902-citibike-tripdata.csv"
tripdata_319 = "Resources/JC-201903-citibike-tripdata.csv"
tripdata_419 = "Resources/JC-201904-citibike-tripdata.csv"
tripdata_519 = "Resources/JC-201905-citibike-tripdata.csv"
tripdata_619 = "Resources/JC-201906-citibike-tripdata.csv"

#2020 Data
tripdata_120 = "Resources/JC-202001-citibike-tripdata.csv"
tripdata_220 = "Resources/JC-202002-citibike-tripdata.csv"
tripdata_320 = "Resources/JC-202003-citibike-tripdata.csv"
tripdata_420 = "Resources/JC-202004-citibike-tripdata.csv"
tripdata_520 = "Resources/JC-202005-citibike-tripdata.csv"
tripdata_620 = "Resources/JC-202006-citibike-tripdata.csv"


#Read in Data
tripdata_119_df = pd.read_csv(tripdata_119)
tripdata_219_df = pd.read_csv(tripdata_219)
tripdata_319_df = pd.read_csv(tripdata_319)
tripdata_419_df = pd.read_csv(tripdata_419)
tripdata_519_df = pd.read_csv(tripdata_519)
tripdata_619_df = pd.read_csv(tripdata_619)

tripdata_120_df = pd.read_csv(tripdata_120)
tripdata_220_df = pd.read_csv(tripdata_220)
tripdata_320_df = pd.read_csv(tripdata_320)
tripdata_420_df = pd.read_csv(tripdata_420)
tripdata_520_df = pd.read_csv(tripdata_520)
tripdata_620_df = pd.read_csv(tripdata_620)
```

```python
tripdata_620_df.head()

all_frames = [tripdata_119_df,tripdata_219_df,tripdata_319_df,tripdata_419_df,tripdata_519_df,tripdata_619_df,tripdata_120_df,tripdata_220_df,tripdata_320_df,tripdata_420_df,tripdata_520_df,tripdata_620_df]
```

```python
result = pd.concat(all_frames, keys=["119","219","319","419","519","619","120","220","320","420","520","620"])
```

```python
result.tail()
```

```python
result.dtypes
```

```python
result[["starttime", "stoptime"]] = result[["starttime", "stoptime"]].apply(pd.to_datetime)

```

```python
result.tail()
result.dtypes
```

```python
result['StartDate']=result["starttime"].dt.date
result['EndDate']=result["stoptime"].dt.date
```

```python
result.dtypes
```

```python
result.head()
```

```python
result['StartTime']=result["starttime"].dt.time
result['StopTime']=result["stoptime"].dt.time
```

```python
result = result.rename(columns={'starttime': 'start_dt_obj', 'stoptime': 'stop_dt_obj'})
```

```python
result.tail()
```

```python
result.to_csv("cleaned_tripdata.csv")
```
